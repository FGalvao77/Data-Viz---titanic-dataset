```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from pylab import rcParams

%matplotlib inline 
plt.style.use('fivethirtyeight')
# plt.style.use('ggplot')
warnings.filterwarnings('ignore')
rcParams['figure.figsize'] = 16, 7
```

```python
# data = pd.read_csv(filepath_or_buffer='titanic.csv')
data = sns.load_dataset(name='titanic')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
plt.plot(data['age'], 
         data['fare'], 
         'o')

plt.title('Scatter plot - Age VS Fare\n', size=20)
plt.xlabel('\nage')
plt.ylabel('fare\n')
plt.show()
```

```python
plt.scatter(
    x=data['age'],
    y=data['fare'],
    c='red'
)

plt.title('Scatter plot: Age VS Fare\n', size=20)
plt.xlabel('\nage')
plt.ylabel('fare\n')
plt.show()
```

```python
color = []

for sex in data['sex']:
    if sex == 'male':
        color.append('blue')
    else:
        color.append('red')

plt.scatter(
    x=data['age'],
    y=data['fare'],
    c=color
)

plt.title('Scatter plot: Age VS Fare', size=20)
plt.xlabel('\nage')
plt.ylabel('fare\n')
plt.show()
```

```python
# instanciando a paleta de cores
color_pal = sns.color_palette()

# visualizando a paleta de cores
color_pal
```

```python
color = []
labels = ['Male', 'Female']

for sex in data['sex']:
    if sex == 'male':
        color.append(color_pal[2])
    else:
        color.append(color_pal[3])

plt.scatter(
    x=data['age'],
    y=data['fare'],
    c=color
)

plt.title('Scatter plot: Age VS Fare', size=20)
plt.xlabel('\nage')
plt.ylabel('fare\n')
plt.show()
```

```python
colors = ['blue' if sex == 'male' else 'red' for sex in data['sex']]

plt.scatter(
    x=data['age'],
    y=data['fare'],
    c=colors
)

plt.title('Scatter plot: Age VS Fare\n', size=20)
plt.xlabel('\nage')
plt.ylabel('fare\n')
plt.show()
```

```python
colors = [color_pal[-1] if sex == 'male' else color_pal[-2] for sex in data['sex']]
classes = ['male', 'female']

fig = plt.scatter(
    x=data['age'],
    y=data['fare'],
    c=colors
)

plt.title('Scatter plot: Age VS Fare\n', size=20)
plt.xlabel('\nage')
plt.ylabel('fare\n')
plt.show()
```

```python
plt.hist(data['age'], bins=50, color=color_pal[1])

plt.title('Histogram: Age\n', size=20)
plt.xlabel('\nage')
plt.ylabel('count\n')
plt.show()
```

```python
data['age'].value_counts().plot(kind='bar')

plt.title('Bar plot: Age\n', size=20)
plt.xticks(rotation=90)
plt.show();
```

```python
data['survived'].value_counts()
```

```python
data['survived'].value_counts().plot(kind='barh')

plt.title('Bar plot: Survived\n', size=20)
plt.xlabel('count')
plt.ylabel('\n0: No | 1: Yes')
plt.show()
```

```python
sns.countplot(data=data)

plt.title('Data count by attributes', size=20)
plt.xlabel('\nattributes')
plt.ylabel('count\n')
plt.show()
```

```python
sns.countplot(x=data['age'])

plt.title('Data count by Age\n', size=20)
plt.xlabel('\nage')
plt.ylabel('count\n')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data=data, y='sex', x='fare', hue='pclass')

plt.title('Box plot: Sex, Fare & Pclass\n', size=20)
plt.xlabel('\nfare')
plt.ylabel('sex\n')
plt.show()
```

```python
sns.boxplot(data=data,
            y='sex',
            x='fare',
            hue='who', 
            orient='h')

plt.title('Box plot: Sex, Fare & Who\n', size=20)
plt.xlabel('\nfare')
plt.ylabel('sex\n')
plt.show()
```

```python
sns.boxenplot(data=data, y='sex', x='fare', hue='pclass')

plt.title('Box plot: Sex, Fare & Pclass\n', size=20)
plt.xlabel('\nfare')
plt.ylabel('sex\n')
plt.show()
```

```python
sns.swarmplot(data=data, y='sex', x='fare', hue='pclass')

plt.title('Box plot: Sex, Fare & Pclass\n', size=20)
plt.xlabel('\nfare')
plt.ylabel('sex\n')
plt.show()
```

```python
sns.stripplot(data=data, x='sex', y='fare', hue='pclass')

plt.title('Box plot: Sex, Fare & Pclass\n', size=20)
plt.xlabel('\nfare')
plt.ylabel('sex\n')
plt.show()
```

```python
corr = data.corr()

sns.heatmap(data=corr,
            annot=True, 
            square=False,
            cmap='Blues',
            fmt='0.4f')

plt.title('Correlation matrix\n', size=20)
plt.show()
```

```python
sns.violinplot(data=data,
               x='sex',
               y='fare',
               hue='who')

plt.show()
```
